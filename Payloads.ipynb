```python
# Dependencies
import pandas as pd
import requests
import json
```

```python
# Reading csv to dataframe
payload_df = pd.read_csv('Resources/payloads.csv')

# Display DF
payload_df
```

```python
# Droping column from DF
payload_df= payload_df.drop(['manufacturers'],axis=1)

# Display DF
payload_df
```

```python
# Copy DF to another DF
payloads_df = payload_df.copy()

# Display DF
payloads_df
```

```python
# Url to retrieve Payload data
url ='https://api.spacexdata.com/v3/payloads'

# Retrive data & print
response = requests.get(url)
response_json = response.json()
print(json.dumps(response_json,indent=4, sort_keys=True))
```

```python
# Adding another column to DF
payloads_df['nationality'] =''

# Add data to newly created column from API data retrived 
for index,row in payloads_df.iterrows():
    for result in response_json:
        if result["payload_id"] == row["name"]:
            if 'nationality' in result:
                payloads_df.loc[index,"nationality"] = result["nationality"]
```

```python
# Display DF
payloads_df
```

```python
#Dropping columns that are not required
payloads_df.drop(payload_df.columns[[0,4,6,7,8]], axis = 1, inplace = True)

# Display DF
payloads_df
```

```python
#Changing the name of columns
payloads_df = payloads_df.rename(columns={'name':'payload_id'})

# Display DF
payloads_df
```

```python
# Fill any empty values for mass_lb with 0
payloads_df['mass_lb'] = payloads_df['mass_lb'].fillna(0)

# Display DF
payloads_df
```

```python
# Replace the empty in Nationality column to Unknown
payloads_df['nationality'] = payloads_df['nationality'].replace('',"Unknown")

# Display DF
payloads_df
```

```python
# Write DF to CSV
payloads_df.to_csv("Clean_CSV/payload.csv", index=False, header=True)
```
