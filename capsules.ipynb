```python
#Importing modules

import pandas as pd
import requests
import json
from pandas import json_normalize
import numpy as np
```

```python
# Reading csv file & store it as DF
file_path_capsules = "Resources/capsules.csv"
capsules_df = pd.read_csv(file_path_capsules)

# Display DF
capsules_df
```

```python
# Rename of column
capsules_df.rename(columns = {'serial':'capsule_serial'}, inplace = True)

# Display preview of DF
capsules_df.head()
```

```python
# Drop column from DF
capsules_df=capsules_df.drop(["capsule_id"],axis =1)

# Display DF
capsules_df
```

```python
# URL for GET requests to retrieve capsules data
url = "https://api.spacexdata.com/v3/capsules"

# Retriving capsule data
capsules_json_response = requests.get(url).json()

# Print the json response
print(json.dumps(capsules_json_response, indent=4, sort_keys=True))
```

```python
# Add two columns to DF
capsules_df['capsule_id']=''
capsules_df['type']=''

# Add capsule_id & capsule type to DF from API response
for index,row in capsules_df.iterrows():
     for result in capsules_json_response:
        if result["capsule_serial"] == row["capsule_serial"]:
            capsules_df.loc[index,"capsule_id"] = result["capsule_id"]
            capsules_df.loc[index,"type"] = result["type"]

# Display DF
capsules_df
```

```python
# Rearrange the columns & display DF
capsules_df2=capsules_df.reindex(columns= ['capsule_serial', 'capsule_id', 'status','type','reuse_count','water_landings','land_landings'])

capsules_df2.head()
```

```python
# Drop any missing values
capsules_df2 = capsules_df2.replace('', np.nan, regex=True)
capsules_df2 = capsules_df2.dropna(how='any')
capsules_df2
```

```python
# Write capsule data to csv
capsules_df2.to_csv("Clean_CSV/capsules_final.csv", index=False, header=True)
```
