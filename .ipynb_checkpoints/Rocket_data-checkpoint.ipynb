```python
# Dependencies
import pandas as pd
import json
import requests
import numpy as np
from pprint import pprint
```

```python
# Read file and store in DataFrame
Rocket_data  = pd.read_csv("Resources/rockets.csv")
```

```python
# API url
url = "https://api.spacexdata.com/v3/rockets"
```

```python
# Retrive API data & print

response = requests.get(url)
response_json = response.json()
print(json.dumps(response_json, indent=4, sort_keys=True))
```

```python
# Create two columns in DF
Rocket_data['id'] =''
Rocket_data['success_rate'] =''

# Read data from API and add to DF 
for index,row in Rocket_data.iterrows():
    for result in response_json:
        if result["rocket_name"] == row["name"]:
            Rocket_data.loc[index,"id"] = result["rocket_id"]
            Rocket_data.loc[index,"success_rate"] = result["success_rate_pct"]

# Display DF
Rocket_data
```

```python
# Rearrange and rename the columns 
Rocket_data_Final = Rocket_data[["id", "name", "type", "active", "country", "company", "height_ft", "diameter_ft", "mass_lb",
                                 "stages","boosters", "cost_per_launch", "success_rate"]]

Rocket_data_Final = Rocket_data_Final.rename(columns={"id": "rocket_id" })

# Display Final DF
Rocket_data_Final
```

```python
# Write DF to CSV
Rocket_data_Final.to_csv ("Clean_csv/rockets.csv", index = False, header=True)
```
