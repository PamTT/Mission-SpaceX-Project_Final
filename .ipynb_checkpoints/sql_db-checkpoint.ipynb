```python
### Store CSV into DataFrame
capsules = pd.read_csv("Clean_CSV/capsules_final.csv")
rockets = pd.read_csv("Clean_CSV/rockets.csv")
payloads = pd.read_csv("Clean_CSV/payload.csv")
launches = pd.read_csv("Clean_CSV/launches_final.csv")
launchpads = pd.read_csv("Clean_CSV/launchpad_final.csv")
bridge = pd.read_csv("Clean_CSV/bridge.csv")
launch_capsule = pd.read_csv("Clean_CSV/launch_capsule.csv")
```

```python
# Check if site_id , rocket_id & payload_id's are present in their respective primary tables(DF's).
# cleanup bridge DF accordingly and store in another DF:

bridge_final = pd.DataFrame(columns=["launch_id", "rocket_id","site_id","payload_id"])

for index,row in bridge.iterrows():
    flag = 0
    launch_id = row["launch_id"]
    rocket_id = row["rocket_id"]
    site_id = row["site_id"]
    payload_id = row["payload_id"]
    
    for index1,row1 in rockets.iterrows():
        if row1["rocket_id"] == rocket_id:
            flag += 1
            
    for index2,row2 in launchpads.iterrows():
        if row2["site_id"] == site_id:
            flag += 1
    
    for index3,row3 in payloads.iterrows():
        if row3["payload_id"] == payload_id:
            flag+=1
              
    if flag == 3:
        bridge_final = bridge_final.append({"launch_id" : launch_id,
                                            "rocket_id" : rocket_id,
                                            "site_id" : site_id,
                                            "payload_id" : payload_id
                                           },ignore_index = True)

# Display Final DF        
bridge_final
```

```python
# Create & establish database connection
engine = create_engine(f'postgresql://{username}:{password}@localhost:5432/Mission_SpaceX')
connection = engine.connect()
```

```python
# Get the table names 
engine.table_names()
```

```python
# Load DataFrames into database
engine.execute("TRUNCATE capsule CASCADE;")
capsules.to_sql(name='capsule', con=engine, if_exists='append', index=False)
pd.read_sql_query('select * from capsule', con=engine).head()
```

```python
# Load DataFrames into database
engine.execute("TRUNCATE rocket CASCADE;")
rockets.to_sql(name='rocket', con=engine, if_exists='append', index=False)
pd.read_sql_query('select * from rocket', con=engine).head()
```

```python
# Load DataFrames into database
engine.execute("TRUNCATE payload CASCADE;")
payloads.to_sql(name='payload', con=engine, if_exists='append', index=False)
pd.read_sql_query('select * from payload', con=engine).head()
```

```python
# Load DataFrames into database
engine.execute("TRUNCATE launches CASCADE;")
launches.to_sql(name='launches', con=engine, if_exists='append', index=False)
pd.read_sql_query('select * from launches', con=engine).head()
```

```python
# Load DataFrames into database
engine.execute("TRUNCATE launchpad CASCADE;")
launchpads.to_sql(name='launchpad', con=engine, if_exists='append', index=False)
pd.read_sql_query('select * from launchpad', con=engine).head()
```

```python
# Load DataFrames into database
engine.execute("TRUNCATE launch_cap CASCADE;")
launch_capsule.to_sql(name='launch_cap', con=engine, if_exists='append', index=False)
pd.read_sql_query('select * from launch_cap', con=engine).head()
```

```python
# Load DataFrames into database
engine.execute("TRUNCATE bridge CASCADE;")
bridge_final.to_sql(name='bridge', con=engine, if_exists='append', index=False)
pd.read_sql_query('select * from bridge', con=engine).head()
```
